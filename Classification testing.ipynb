```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate two-dimensional data
n_samples = 1000

# Class 0: samples from a 2D normal distribution
n_class_0 = n_samples // 2
mean_0 = [0, 0]
cov_0 = [[1, 0], [0, 1]]
class_0 = np.random.multivariate_normal(mean_0, cov_0, n_class_0)

# Class 1: samples from a different 2D normal distribution
n_class_1 = n_samples - n_class_0
mean_1 = [2, 2]
cov_1 = [[1.5, 0], [0, 1.5]]
class_1 = np.random.multivariate_normal(mean_1, cov_1, n_class_1)

# Combine the data
X = np.vstack((class_0, class_1))
y = np.hstack((np.zeros(n_class_0), np.ones(n_class_1)))

# Create a DataFrame
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
df['Target'] = y.astype(int)

# Save to CSV
df.to_csv('classification_dataset.csv', index=False)

print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
print("\nClass distribution:")
print(df['Target'].value_counts(normalize=True))

# Visualize the data
plt.figure(figsize=(10, 8))
plt.scatter(df[df['Target'] == 0]['Feature_1'], df[df['Target'] == 0]['Feature_2'], label='Class 0', alpha=0.6)
plt.scatter(df[df['Target'] == 1]['Feature_1'], df[df['Target'] == 1]['Feature_2'], label='Class 1', alpha=0.6)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Generated Classification Dataset')
plt.legend()
plt.show()
```
